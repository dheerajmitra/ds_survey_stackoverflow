```python
def color_plot(data_frame,title_choose):
    """ This function is  used to crete a bar chart with different colors"""
    data_frame.plot(kind="bar");
    plt.title(title_choose);
```

```python
job_profile=((bootcamp_df['TimeAfterBootcamp'].value_counts()/bootcamp_df.shape[0])/df.shape[0])
color_plot(job_profile,"Bootcamp job profiles")
```
